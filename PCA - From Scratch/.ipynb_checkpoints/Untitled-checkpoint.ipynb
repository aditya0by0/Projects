### PCA - From Scratch

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import animation
```

```python
X = np.loadtxt('pca_data',delimiter=',') # loadind data to numpy array
X[:5] 
```

```python
# ---------- Plotting the data ----------

plt.figure(figsize=(3,3))
plt.plot(X[:,0], X[:,1], 'bo', ms=10, mec='k', mew=1 )
plt.plot(np.mean(X,axis=0)[0],np.mean(X,axis=0)[1], 'r+', ms=10, mew=2) # mean of the data
plt.axis([0.5, 6.5, 2, 8])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
#plt.title('PCA Data')
plt.show()
```

#### Normalization : 

Normalization means to make mean of data to 0 and scale by standard deviation

```python
# ---------- Normalizing the data -----------

X_norm = np.zeros(X.shape)
mean = np.mean(X,axis=0)
std = np.std(X, axis=0, ddof=1) # sample variance (std)
X_norm = (X - mean)/std # normalize and scale

plt.figure(figsize=(3,3))
plt.plot(X_norm[:,0], X_norm[:,1], 'bo', ms=10, mec='k', mew=1 )
# mean of the data
norm_mean =  np.mean(X_norm,axis=0)
plt.plot( norm_mean[0], norm_mean[1], 'r+', ms=10, mew=2) 
plt.axis([-3, 3, -3, 3])
plt.show()
```

#### Covariance Matrix :

Matrix which gives variance of each feature on its diagonal elements and covariance on its non-diagonal elements


```python
# ---------- Covariance Matrix of Normalized data ------------------

Covariance_matrix = (1/(X_norm.shape[0])) * np.dot(X_norm.T,X_norm) # nxm - mxn
Covariance_matrix # nxn - covariance matix of features of the data
```

#### Covariance Matrix :

Matrix which gives variance of each feature on its diagonal elements and covariance on its non-diagonal elements


```python
# ---------- Covariance Matrix of Normalized data ------------------

Covariance_matrix = (1/(X_norm.shape[0])) * np.dot(X_norm.T,X_norm) # nxm - mxn
Covariance_matrix # nxn - covariance matix of features of the data
```

we can see that our covariance is symmetric matrix

##### 1. Find the Eigen values and Eigen vectors of our Covariance Matrix

```python
# --------- Finding eigen values and vectors --------
def find_eigens(Cov_matrix):
    eigen_values, eigen_vectors = np.linalg.eig(Cov_matrix) # returns tuple of eigen values and unit eigen vectors
    sort_indices = np.argsort(eigen_values)[::-1] # sort the indices of  eigen value from large to small
    eigen_values = eigen_values[sort_indices] # sort the eigen values
    eigen_vectors = eigen_vectors[:,sort_indices] # sort eigen column vectors in same order
    return eigen_values, eigen_vectors

lambda_, U = find_eigens(Covariance_matrix) 
U
```

```python
def cal_var_pc(array):
    
    total_variance = np.sum(array)
    var_in_pc = []
    for i in array:
        var_4_pc = round(i/total_variance,2)
        var_in_pc.append(var_4_pc)
        
    return var_in_pc

pca_variance = cal_var_pc(lambda_)
```

```python
%matplotlib inline
plt.bar(['pc1','pc2'], pca_variance)
plt.show()
```

```python
fig,ax =  plt.subplots(figsize=(5,5),dpi=100)
ax.plot(X_norm[:, 0], X_norm[:, 1], 'bo', ms=10, mec='k', mew=1)

ax.plot( norm_mean[0], norm_mean[1], 'r+', ms=10, mew=2) 


for i in range(2):
    ax.arrow(norm_mean[0], norm_mean[1],  lambda_[i]*U[0,i], lambda_[i] * U[1,i],
             head_width=0.25,head_length=0.2, fc='k', ec='k', lw=2, zorder=1000)
    
plt.axis([-3, 3, -3, 3])
plt.show()
```

##### 3. Orthonormalize our transformed vectors

```python
fig,ax =  plt.subplots(figsize=(5,5),dpi=100)
ax.plot(X_norm[:, 0], X_norm[:, 1], 'bo', ms=10, mec='k', mew=1)

ax.plot( norm_mean[0], norm_mean[1], 'r+', ms=10, mew=2) 

ax.arrow(norm_mean[0], norm_mean[1],  lambda_[0]*U[0,0], lambda_[0] * U[1,0],
             head_width=0.25,head_length=0.2, fc='k', ec='k', lw=2, zorder=1000)

Z = np.dot(X_norm,U[:,0:1])
Proj = U[:,0:1].T * Z
for xnorm, proj in zip(X_norm, Proj):
    ax.plot([xnorm[0], proj[0]], [xnorm[1], proj[1]], '--k', lw=1)

ax.plot([-3.5 * U[0,0], 3.5 * U[0,0]] ,[-3.5 * U[1,0],3.5 * U[1,0]], lw = 0.5  ,color='black' )
plt.axis([-3, 3, -3, 3])
plt.show()
```

```python
def project_onto(X,U,K):
    Z = np.dot(X,U[:,:K])
    return Z
Z = project_onto(X_norm,U,1) 
plt.figure(figsize=(3,1),dpi=150)
plt.plot(Z,np.zeros(len(Z)),'bo', ms=8, mec='k', mew=0.5)

plt.show()
```

```python
def recover_back(Z,U,K):
    X_rec = np.dot(Z,U[:,:K].T)
    return X_rec

X_rec = recover_back(Z,U,1)
fig, ax = plt.subplots(figsize=(5, 5),dpi=100)
ax.plot(X_norm[:, 0], X_norm[:, 1], 'bo', ms=10, mec='k', mew=1)

plt.axis([-3, 3, -3, 3])

# Draw lines connecting the projected points to the original points
ax.plot(X_rec[:, 0], X_rec[:, 1], 'ro', mec='r', mew=2, mfc='none')
for xnorm, xrec in zip(X_norm, X_rec):
    ax.plot([xnorm[0], xrec[0]], [xnorm[1], xrec[1]], '--k', lw=1)

plt.legend()
plt.show()
```
